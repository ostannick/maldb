```python
def Gaussian(x, intensity, center, stdev):
    
    return intensity * np.exp(-((x - center)**2 / (2 * stdev**2)))
```

```python
intensities = np.random.uniform(10, 100, 22)
masses = np.random.uniform(650, 4200, 22)
stdevs = np.random.uniform(0.3, 1, 22)

y_series = np.random.uniform(0, 1, x_series.shape)

for x in range(0, len(intensities)):
    
    y_series += Gaussian(x_series, intensities[x], masses[x], stdevs[x])
```

```python
plt.plot(x_series, y_series)
```
